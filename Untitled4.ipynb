```python
# question 1
import numpy as np
import scipy.stats as stats

def gibbs_sampler(n, m, s, num_samples=1000):
    # Generate observed data
    x = stats.norm(loc=m, scale=s).rvs(size=n)
    
    # Initialize parameters
    theta = np.mean(x)  # Start with sample mean
    tau = 1.0 / np.var(x)  # Start with precision as inverse variance
    
    theta_samples = []
    tau_samples = []
    
    for _ in range(num_samples):
        # Sample from p(theta | tau, x), which is normal
        theta_mean = np.mean(x)
        theta_var = 1.0 / (n * tau)
        theta = stats.norm(loc=theta_mean, scale=np.sqrt(theta_var)).rvs()
        
        # Sample from p(tau | theta, x), which is gamma
        shape = n / 2
        scale = 1.0 / (0.5 * np.sum((x - theta) ** 2))  # Inverse of rate
        tau = stats.gamma(a=shape, scale=scale).rvs()
        
        # Store samples
        theta_samples.append(theta)
        tau_samples.append(tau)
    
    return np.array(theta_samples), np.array(tau_samples)

# Example usage
n, m, s = 30, 5, 2
theta_samples, tau_samples = gibbs_sampler(n, m, s)
print(f"Mean of sampled theta: {np.mean(theta_samples)}")
print(f"Mean of sampled tau: {np.mean(tau_samples)}")
```

```python
# question 2
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# --- Step 1: Define Data and Priors ---
n = 30  # Number of observations
x = stats.norm(loc=0, scale=1).rvs(size=n)  # Generate synthetic data

# Prior hyperparameters
theta_0 = 0  # Mean prior mean
tau_0 = 1  # Mean prior precision (inverse variance)
alpha = 2  # Shape parameter for gamma prior
lambda_ = 2  # Rate parameter for gamma prior

# Gibbs Sampler Setup
num_samples = 10000
burn_in = 500

theta_samples = np.zeros(num_samples)
tau_samples = np.zeros(num_samples)

theta = np.mean(x)  # Initialize theta
tau = 1.0  # Initialize tau

# --- Step 2: Gibbs Sampling ---
for i in range(num_samples):
    # Sample tau from Gamma( (alpha + n)/2 , (lambda + sum (x_i - theta)^2)/2 )
    alpha_post = (alpha + n) / 2
    lambda_post = (lambda_ + np.sum((x - theta) ** 2)) / 2
    tau = stats.gamma.rvs(alpha_post, scale=1/lambda_post)
    
    # Sample theta from N( (n*xÌ„*tau + theta_0*tau_0) / (n*tau + tau_0) , 1 / (n*tau + tau_0) )
    mean_post = (n * np.mean(x) * tau + theta_0 * tau_0) / (n * tau + tau_0)
    var_post = 1 / (n * tau + tau_0)
    theta = stats.norm.rvs(mean_post, np.sqrt(var_post))
    
    theta_samples[i] = theta
    tau_samples[i] = tau

# --- Step 3: Plot Results ---
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(theta_samples[burn_in:], bins=50, density=True, alpha=0.7, color='blue')
ax[0].set_title("Posterior Distribution of $\\theta$")
ax[0].set_xlabel("$\\theta$")
ax[0].set_ylabel("Density")

ax[1].hist(tau_samples[burn_in:], bins=50, density=True, alpha=0.7, color='green')
ax[1].set_title("Posterior Distribution of $\\tau$")
ax[1].set_xlabel("$\\tau$")
ax[1].set_ylabel("Density")

plt.show()
```

```python
# question 3
# Multinomial-Dirichlet Derivation

# 1. Import necessary libraries
import numpy as np
from scipy.stats import dirichlet, multinomial
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 2. Define the model setup in a Markdown cell:
"""
### Multinomial-Dirichlet Model Setup

1. **Likelihood**: The likelihood of observing counts \(x_1, x_2, \ldots, x_k\) given the probabilities \(\boldsymbol{\theta} = (\theta_1, \theta_2, \ldots, \theta_k)\) is:
   \[
   P(x_1, x_2, \ldots, x_k \,|\, \boldsymbol{\theta}) = \frac{n!}{x_1! x_2! \cdots x_k!} \prod_{i=1}^k \theta_i^{x_i},
   \]
   where \(n = \sum_{i=1}^k x_i\) is the total number of trials.

2. **Prior**: The Dirichlet prior for \(\boldsymbol{\theta}\) is:
   \[
   P(\boldsymbol{\theta}) = \frac{\Gamma\left(\sum_{i=1}^k \alpha_i\right)}{\prod_{i=1}^k \Gamma(\alpha_i)} \prod_{i=1}^k \theta_i^{\alpha_i - 1},
   \]
   where \(\alpha_i > 0\) are the prior parameters.

3. **Posterior**: Using Bayes' theorem, the posterior is proportional to the product of the prior and likelihood:
   \[
   P(\boldsymbol{\theta} \,|\, \boldsymbol{x}) \propto P(\boldsymbol{x} \,|\, \boldsymbol{\theta}) P(\boldsymbol{\theta}).
   \]

4. Substituting the likelihood and prior:
   \[
   P(\boldsymbol{\theta} \,|\, \boldsymbol{x}) \propto \prod_{i=1}^k \theta_i^{x_i} \prod_{i=1}^k \theta_i^{\alpha_i - 1}
   = \prod_{i=1}^k \theta_i^{x_i + \alpha_i - 1}.
   \]

5. The posterior is also a Dirichlet distribution:
   \[
   P(\boldsymbol{\theta} \,|\, \boldsymbol{x}) = \text{Dirichlet}(\boldsymbol{\alpha}' = \boldsymbol{\alpha} + \boldsymbol{x}),
   \]
   where \(\alpha_i' = \alpha_i + x_i\).
"""
# 3. Example of a Multinomial-Dirichlet model
# Define parameters for prior and observed data

# Prior parameters (Dirichlet prior concentration parameters)
alpha = np.array([2, 2, 2])  # Prior belief for 3 categories

# Observed data (Multinomial counts)
data = np.array([5, 3, 7])  # Counts for the same 3 categories

# Update posterior parameters
posterior_alpha = alpha + data

# 4. Draw samples from the posterior
# Posterior is a Dirichlet distribution with updated parameters
n_samples = 5000
posterior_samples = np.random.dirichlet(posterior_alpha, size=n_samples)

# 5. Visualize the posterior distribution
# Plot posterior samples for each category
plt.figure(figsize=(10, 6))
for i in range(len(alpha)):
    sns.histplot(posterior_samples[:, i], bins=30, kde=True, label=f"Category {i + 1}", alpha=0.7)

plt.title("Posterior Samples for Multinomial-Dirichlet Model")
plt.xlabel("Probability")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

## Q1: Posterior Distribution for β
Likelihood:
p(y | X, β, σ²) ∝ exp[ -1/(2σ²) (y - Xβ)ᵀ(y - Xβ) ]

Represents the normal distribution of the data y given X, β, and σ².

Prior:
β ~ N(β₀, Σβ)

Assumes β follows a multivariate normal distribution with mean β₀ and covariance Σβ.

Posterior Derivation:

Combine Likelihood and Prior:

The posterior is proportional to likelihood × prior.

Exponent terms from the likelihood and prior are summed.

Posterior Precision Matrix:
Σ_post⁻¹ = (1/σ²) XᵀX + Σβ⁻¹

(1/σ²) XᵀX: Precision from the likelihood.

Σβ⁻¹: Precision from the prior.

Posterior Mean:
μ_post = Σ_post [ (1/σ²) Xᵀy + Σβ⁻¹β₀ ]

Weighted average of prior mean β₀ and data-driven term Xᵀy.

Final Posterior:
β | y, X, σ² ~ N(μ_post, Σ_post)


### Q1.2: Posterior Distribution for σ²
Prior:
σ² ~ IG(α*, β*)

Inverse-gamma prior with shape α* and scale β*.

Posterior Derivation:

Likelihood for σ²:
p(y | X, β, σ²) ∝ (σ²)^(-n/2) exp[ -1/(2σ²) (y - Xβ)ᵀ(y - Xβ) ]

Depends on the residual sum of squares (RSS).

Combine with Prior:

Multiply by inverse-gamma prior: (σ²)^(-α* -1) exp(-β*/σ²).

Updated Parameters:

Shape: α* + n/2

Scale: β* + ½(y - Xβ)ᵀ(y - Xβ)

Final Posterior:
σ² | y, X, β ~ IG(α* + n/2, β* + ½RSS)

### Q1 Extra: Connection to Ridge and Lasso
Ridge Regression (Normal Prior):

Log-Posterior:
-1/(2σ²)‖y - Xβ‖² - ½‖β‖²

-½‖β‖² acts as an L2 penalty term.

Analogous to ridge regression loss: ‖y - Xβ‖² + λ‖β‖².

Lasso (Laplace Prior):

Log-Posterior:
-1/(2σ²)‖y - Xβ‖² - λ‖β‖₁

-λ‖β‖₁ acts as an L1 penalty term.

Analogous to lasso regression loss: ‖y - Xβ‖² + λ‖β‖₁.



```python
# question2 
try:
    import pymc as pm
except ImportError:
    !pip install pymc arviz scikit-learn pandas numpy
    import pymc as pm

import numpy as np
import pandas as pd
import arviz as az
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler

# Load data
data = fetch_california_housing()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Standardize features and target
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = (y - y.mean()) / y.std()

# Define and run Bayesian model
with pm.Model() as model:
    # Priors
    p = X_scaled.shape[1]
    betas = pm.Normal('betas', mu=0, sigma=1, shape=p)
    sigma = pm.HalfNormal('sigma', sigma=1)  # Non-inverse-gamma prior
    
    # Likelihood
    mu = pm.math.dot(X_scaled, betas)
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed=y_scaled)
    
    # MCMC Sampling
    trace = pm.sample(1000, tune=1000, chains=4, cores=1)

# Diagnostics and posterior analysis
az.summary(trace, hdi_prob=0.95)  # Parameter summary
az.plot_trace(trace, var_names=['betas', 'sigma'])  # Trace plots
```

```python
import numpy as np
import pymc as pm
import pytensor.tensor as pt  # ✅ Use Aesara tensor operations

# 🔹 Generate synthetic data
np.random.seed(42)
X = np.random.randn(100, 2)  # 100 samples, 2 predictors
beta_true = np.array([3, -2])
y = X @ beta_true + np.random.normal(0, 1, size=100)

# Introduce outliers
outlier_indices = np.random.choice(100, size=10, replace=False)
y[outlier_indices] += np.random.normal(0, 10, size=10)

# 🔹 Define Bayesian Model
with pm.Model() as model:
    # Priors for regression coefficients
    beta = pm.Normal("beta", mu=0, sigma=10, shape=2)
    
    # Global precision
    tau = pm.Gamma("tau", alpha=2, beta=2)
    
    # Local precision terms for robustness
    tau_i = pm.Gamma("tau_i", alpha=2, beta=2, shape=len(y))
    
    # ✅ Use Aesara's tensor operations instead of NumPy
    sigma_i = pm.Deterministic("sigma_i", 1 / pt.sqrt(tau * tau_i))
    
    # ✅ Use Aesara's pt.dot instead of np.dot
    y_obs = pm.Normal("y_obs", mu=pt.dot(X, beta), sigma=sigma_i, observed=y)
    
    # 🔹 Sample from posterior
    trace = pm.sample(1000, return_inferencedata=True)

# 🔹 Identify Outliers
import arviz as az
az.plot_posterior(trace, var_names=["tau_i"])



```
